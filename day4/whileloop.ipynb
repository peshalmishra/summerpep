```python
n = int(input())
n1 = n
sum = 0

while n1 > 0:
    d = n1 % 10
    sum = sum + d
    n1 = n1 // 10

print(sum)
```

```python
n = int(input())
sum = 0

while n != 0:
    sum += n
    n = int(input())
    
print(sum)
```
