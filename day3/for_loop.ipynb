# For Loop Examples
This notebook contains 5 examples using `for` loops in Python.

```python
# Example 1: Iterate over a list
colors = ["red", "green", "blue"]
for color in colors:
    print(color)
```

```python
# Example 2: Iterate over a range
for i in range(5):
    print(i)
```

```python
# Example 3: Sum numbers with a loop
total = 0
for num in range(1, 6):
    total += num
print("Total:", total)
```

```python
# Example 4: Loop with string formatting
for i in range(1, 4):
    print(f"Step {i} completed")
```

```python
# Example 5: Nested loops
for i in range(1, 4):
    for j in range(1, 3):
        print(f"i={i}, j={j}")
    print()
```
