```python
# Print 5 stars on one line
print("*" * 5)
```

```python
# Print a 5x5 square of stars
for _ in range(5):
    print("*" * 5)
```

```python
# Print descending star rows from 5 to 1
for count in range(5, 0, -1):
    print("*" * count)
```

```python
# Print descending number sequences starting from 5
for start in range(5, 0, -1):
    print("".join(str(num) for num in range(start, 0, -1)))
```

```python
# Print an increasing number triangle
for row in range(1, 6):
    print("".join(str(num) for num in range(1, row + 1)))
```

```python
# Print rows of increasing numbers with decreasing length
for row in range(5, 0, -1):
    print("".join(str(num) for num in range(1, row + 1)))
```

```python
# Print a square where each row repeats the row number 5 times
for row in range(1, 6):
    print(str(row) * 5)
```

```python
# Print a triangle of uppercase letters from A
for row in range(1, 6):
    print("".join(chr(65 + col) for col in range(row)))
```

```python
# Print a triangle of repeated numbers: 1, 22, 333, 4444, 55555
for row in range(1, 6):
    print(str(row) * row)
```
