# Conditional Statements
This notebook shows 5 examples of Python conditional statements.

```python
# Example 1: if statement
x = 10
if x > 5:
    print("x is greater than 5")
```

```python
# Example 2: if-else statement
n = 7
if n % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
# Example 3: if-elif-else chain
score = 85
if score >= 90:
    grade = "A"
elif score >= 80:
    grade = "B"
else:
    grade = "C"
print("Grade:", grade)
```

```python
# Example 4: nested conditions
age = 18
has_ticket = True
if age >= 18:
    if has_ticket:
        print("Entry allowed")
    else:
        print("Ticket required")
else:
    print("Too young")
```

```python
# Example 5: conditional expressions
temperature = 20
status = "Warm" if temperature > 18 else "Cold"
print(status)
```
